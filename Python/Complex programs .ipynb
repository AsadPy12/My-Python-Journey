#### 1- - Average Calculator:
<font color = purple> Write a function calculate_average that takes a list of numbers as input and returns the average of those numbers.</font>

```python
nums = [1, 2, 3, 4, 5, 6, 7, 8, 65, 23, 2, 3, 4, 76, 7, 10]
def calc_avg():
    total = sum(nums)
    avg = total / len(nums) 
    
    return avg
print(f' The average of a numbers given in a list is {calc_avg()}')
```

```python
class Student:
    def __init__(self, name):
        self.name = name
    
    def calc_avg(self):
        return sum(self.marks) / len(self.marks)
    
    def calc_grade(self, marks):
        average_marks = self.calc_avg()
        if average_marks >= 90:
            return 'A+'
        elif average_marks >= 80:
            return 'A'
        elif average_marks >= 70:
            return 'B'
        elif average_marks >= 60:
            return 'C'
        elif average_marks >= 50:
            return 'D'
        elif average_marks >= 40:
            return 'E'
        else:
            return '!!!Fail!!!'

def main():
    name = input("Enter Your Name: ")
    student = Student(name)
    
    num_grades = int(input("Enter Your Numbers: "))
    
    grades = {}
    for i in range(num_grades):
        grade = int(input("Enter grade (i + 1): "))
        grades.update(grade)
        
    grade = student.calc_grade(grades)
    print(f"{student.name}'s grade is: {grade}")
    
if __name__ == "__main__":
    main()
```

### Assignment: Modern Banking Application

```python
class BankAccount():
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance
    
    def credit(self, amount):
        if amount > 0:
            self.balance += amount 
            print(f' ${amount} is credited into your Account. Now your Balance is {self.balance}')
        else:
            print("Invalid credit amount.")
            
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f' ${amount} is Withdrawed into your Account. New Balance is ${self.balance}')
        else:
            print('Insufficient funds')
            
    def check_balance(self):
        print(f'Balance for {self.name} is ${self.balance}')
        
        
def main():
    print('*_*_*_*_*_*_*_*_*-*-*_*_*_*_*_*_*_*_*')
    print("Welcome to the Asad's Bank Account System!")
    name = input("Enter your Name: ")
    account = BankAccount(name)
    
    while True:
        print('\nBank Menu:')
        print('1- Credit')
        print('2- Withdraw')
        print('3- Check Balance')
        print('4- Exit')
        choice = input("Enter your choice between(1-4): ")

        if choice == '1':
            amount = float(input("Enter the Amount: "))
            account.credit(amount)
        elif choice == '2':
            amount = float(input("Enter the Amount: "))
            account.withdraw(amount)
        elif choice == '3':
            account.check_balance()
        elif choice == '4':
            print('Exiting the Account, GoodBye!!')
            print('*_*_*_*_*_*_*_*_*-*-*_*_*_*_*_*_*_*_*')
            break
        else:
            print("Invalid option. Please choose a valid option between 1-4::")
        
if __name__ == '__main__':
    main()
```

#### Part1, Assignment: Managing Attendance in a Classroom

You're a teacher in charge of a classroom, and it's your responsibility to keep track of students' attendance. You decide to utilize a simple attendance register program to streamline the process.

<font color = darkred>1- Marking Attendance:</font>                       
<font color = darkgreen>Every morning, as students enter the classroom, you use the program to mark their attendance. You select the option to mark attendance and input each student's name one by one. If a student is already marked present, the program notifies you, ensuring no duplicate entries are made.</font> 

<font color = darkred>2- Displaying Attendance:</font>                    
<font color = redgreen>During the day, you might need to refer to the attendance record for various purposes. You select the option to display attendance, and the program presents you with a list of all students along with their attendance status—whether they are present or absent.</font> 

<font color = darkred>3- End of the Day Routine:</font>                   
<font color = green>At the end of the school day, you review the attendance record one last time. You ensure that all students who were present have been marked so and that no mistakes have been made in the entries.</font> 

<font color = darkred>4- Exiting the Program:</font>                       
<font color = redgreen>After completing your tasks, you select the option to exit the program. The program gracefully terminates, allowing you to attend to other responsibilities.</font> 

This scenario illustrates how the attendance register program simplifies the process of managing attendance in the classroom, offering efficiency and accuracy in tracking students' presence.

```python
class Attendance_Reg:
    def __init__(self):
        self.attendance = {}
        
    def marking_attendance(self, student_name):
        if student_name in self.attendance:
            print(f'{student_name} attendance is already marked as present')
        else:
            self.attendance[student_name] = True
            print(f'{student_name} attendance is marked as present.')
            
    def displaying_attendance(self):
        for student, present in self.attendance.items():
            status = "Present" if present else "Absent"
            print(f'{student}: {status}')
            
def main():
    register = Attendance_Reg()
    
    while True:
        print('\n1- Mark Attendance')
        print('2- Display Attendance')
        print('3- Exiting the System')
        choice = input("Enter the choice between (1-3): ")
        
        if choice == '1':
            student_name = input("Enter the student name: ")
            register.marking_attendance(student_name)
        elif choice == '2':
            register.displaying_attendance()
        elif choice == '3':
            print('Exiting program')
            break
        else:
            print('Invalid choice. Please try again with a valid number')

if __name__ == '__main__':
    main()
```

```python
class Attendance():
    def __init__(self):
        self.attendance = {}
    
    def add_student(self, name):
        self.attendance[name] = "Absent"
        
    def mark_present(self, name):
        if name in self.attendance:
            self.attendance[name] = 'Present'
        else:
            print(f'{name} is not in the register')
            
    def mark_absent(self, name):
        if name in self.attendance:
            self.attendance[name] = 'Absent'
        else:
            print(f'{name} is not in the register')
            
    def view_attend(self):
        for name, status in self.attendance.items():
            print(f'{name}: {status}')
            
def main():
    
    print('\n===============================================')
    print('Welcome to the ABC school attendance system...')
    print('===============================================')
    
    print('*************************')
    print('--------------------------')
    print('This page is for class 9')
    print('--------------------------')
    print('*************************')
    
    print("=====H*E*L*L*O=====)")
    register = Attendance()
    num_students = int(input('Enter the nums of students in the class: '))
    
    for _ in range(num_students):
        name = input('Enter the student name: ')
        register.add_student(name)
        
    while True:
        print("\n1. Mark Present")
        print("2. Mark Absent")
        print("3. View Attendance")
        print("4. Exit")
        choice = input('Enter your choice: ')
        
        if choice == '1':
            name = input('Enter student name to mark present: ')
            register.mark_present(name)
        elif choice == '2':
            name = input('Enter student name to mark absent: ')
            register.mark_absent(name)
        elif choice == '3':
            register.view_attend()
        elif choice == '4': 
            print('********************************')
            print('Exiting the attendance system: ')
            print('********************************')
            print('Goodbyee.....')
            print('================================')
            print('________________________________')
            break
        else:
            print('Inappropriate choice. Enter valid choice.')
    
if __name__ == '__main__':
    main()
```

### Assignment: Rock, Paper, Scissor Game

```python
## Rock Paper Scissor (RPS) Game:

import random


print('=================================================')
print("***************Welcome to RPS Game***************")
print('=================================================')

print("")
name = input("Enter your Name here: ")
user_score = 0
ties = 0 
comp_score = 0

print("""-----------------------------------
Game Rules:
1- Rock vs Scissor --> Rock wins
2- Rock vs Paper --> Paper wins
3- Paper vs Scissor --> Scissor
-----------------------------------""")
print("""-------------
Choices Are:
1- Rock
2- Paper
3- Scissor
-------------
""")

while True:
    choice = int(input("Enter your choice between(1-3): "))

    while choice > 3 or choice < 1:
        choice = int(input("Please Enter a valid choice: "))

    if choice == 1:
        user_choice = "Rock"
    elif choice == 2:
        user_choice = "Paper"
    elif choice == 3:
        user_choice = "Scissor"
    print('==================================')
    print("The user choice is:", user_choice)
    print('===================================')
    print("Now its Computer's turn.....")
    print("-----------------------------------")
    computer = random.randint(1, 3)

    if computer == 1:
        com_choice = "Rock"
    elif choice == 2:
        com_choice = "Paper"
    elif choice == 3:
        com_choice = "Scissor"

    print('**************************************')
    print('The computer choice is:', com_choice)
    print('**************************************')
    if (user_choice == "Paper" and com_choice == "Rock") or (user_choice == "Rock" and com_choice == "Paper"):
        result = "Paper"
    elif(user_choice == "Scissor" and com_choice == "Rock") or (user_choice == "Rock" and com_choice == "Scissor"):
        result = "Rock"
    elif(user_choice == com_choice):
        print("GAME TIES")
        result = "Tie"
    else:
        result ="Scissor"

    if result == "Tie":
        ties += 1
    elif result == user_choice:
        user_score += 1
        print(f"{name} wins")
    else:
        comp_score += 1
        print("Computer's wins")
        
    print('$$$$$$$$*****$$$$$$$$$$$$*****$$$$$$')
    print("Scores Are:")
    print('------------------------------------')
    print(f'{name} score is: {user_score}')
    print('====================================')
    print(f"Computer's score is: {comp_score}")
    print('************************************')
    print(f'Match are ties: {ties}')    
    print('$$$$$$$$*****$$$$$$$$$$$$*****$$$$$$')
    print('=====================================')
    print('-------------------------------------')
    print('-------------------------------------')

    repeat = input("Do you want to play again,(NO or No) is for quit..: ")
    if repeat == "No" or repeat == "NO":
        break
print()
print("=================================================")
print("Game is Over!")
print("-------------------------------------------------")
```
