### Assignment: Modern Banking Application

```python
class BankAccount():
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance
    
    def credit(self, amount):
        if amount > 0:
            self.balance += amount 
            print(f' ${amount} is credited into your Account. Now your Balance is {self.balance}')
        else:
            print("Invalid credit amount.")
            
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f' ${amount} is Withdrawed into your Account. New Balance is ${self.balance}')
        else:
            print('Insufficient funds')
            
    def check_balance(self):
        print(f'Balance for {self.name} is ${self.balance}')
        
        
def main():
    print('*_*_*_*_*_*_*_*_*-*-*_*_*_*_*_*_*_*_*')
    print("Welcome to the Asad's Bank Account System!")
    name = input("Enter your Name: ")
    account = BankAccount(name)
    
    while True:
        print('\nBank Menu:')
        print('1- Credit')
        print('2- Withdraw')
        print('3- Check Balance')
        print('4- Exit')
        choice = input("Enter your choice between(1-4): ")

        if choice == '1':
            amount = float(input("Enter the Amount: "))
            account.credit(amount)
        elif choice == '2':
            amount = float(input("Enter the Amount: "))
            account.withdraw(amount)
        elif choice == '3':
            account.check_balance()
        elif choice == '4':
            print('Exiting the Account, GoodBye!!')
            print('*_*_*_*_*_*_*_*_*-*-*_*_*_*_*_*_*_*_*')
            break
        else:
            print("Invalid option. Please choose a valid option between 1-4::")
        
if __name__ == '__main__':
    main()
```

```python
class Attendance_Reg:
    def __init__(self):
        self.attendance = {}
        
    def marking_attendance(self, student_name):
        if student_name in self.attendance:
            print(f'{student_name} attendance is already marked as present')
        else:
            self.attendance[student_name] = True
            print(f'{student_name} attendance is marked as present.')
            
    def displaying_attendance(self):
        for student, present in self.attendance.items():
            status = "Present" if present else "Absent"
            print(f'{student}: {status}')
            
def main():
    register = Attendance_Reg()
    
    while True:
        print('\n1- Mark Attendance')
        print('2- Display Attendance')
        print('3- Exiting the System')
        choice = input("Enter the choice between (1-3): ")
        
        if choice == '1':
            student_name = input("Enter the student name: ")
            register.marking_attendance(student_name)
        elif choice == '2':
            register.displaying_attendance()
        elif choice == '3':
            print('Exiting program')
            break
        else:
            print('Invalid choice. Please try again with a valid number')

if __name__ == '__main__':
    main()
```

```python
def create_file(filename):
    try:
         with open(filename, 'w') as file:
            print('File created successfully.')
    except:
        print('Error: Could not create File.')
        
def read_file(filename):
    try:
        with open(filename, 'r')as file:
            print("File contents:")
            print(file.read())
    except:
        print('Error: File not found.')
    
def write_file(filename):
    try:
        with open(filename, 'w') as file:
            content = input('Enter content to write: ')
            file.write(content)
            print('content written successfully.')
    except:
        print("Error: Could not write to file")
        
def append_file(filename):
    try:
        with open(filename, 'a') as file:
            content = input("Enter the content to append: ")
            file.write(content)
            print("Content upgraded successfully.")
    except:
        print("Error: Upgrading Not possible.")
        
        
def main():
    while True:
        print("\nMenu: ")
        print("1. Create a new file")
        print("2. Read a file")
        print("3. Write to a file")
        print("4. Append to a file")
        print("5. Exit")
        
        choice = int(input('Enter your choice: '))
        
        if choice == 1:
            filename = input("Enter filename: ")
            create_file(filename)
        elif choice == 2:
            filename = input("Enter filename: ")
            read_file(filename)
        elif choice == 3:
            filename = input("Enter filename: ")
            write_file(filename)
        elif choice == 4:
            filename = input("Enter filename: ")
            append_file(filename)
        elif choice == 5:
            print("Exiting the program, GoodBye!!")
            break
        else:
            print("Invalid choice. Enter the valid option.")
            
if __name__ == '__main__':
    main()
```

#### Assignment : Student Grade Calculator
<font color = red> **Write a Python program that calculates the average grade of students and determines their overall performance.**</font>

<font color = fuchsia>Requirements:</font>

<font color = green>1- The program should prompt the user to enter the number of students.

<font color = green>2- For each student, the program should prompt the user to enter the name and grades (assume only three subjects: Math, Science, and English).

<font color = green>3- Calculate the average grade for each student.</font>

<font color = green>4- Determine the overall performance of each student based on the average grade:</font>
- If the average grade is >= 90, the performance is "Excellent".
- If the average grade is >= 80 and < 90, the performance is "Good".
- If the average grade is >= 70 and < 80, the performance is "Average".
- If the average grade is < 70, the performance is "Below Average".

<font color = green>5- Print out the name, average grade, and performance of each student.</font>

<font color = blue>**Constraints:**</font>

- Assume all input grades are integers.
- Assume names do not contain any special characters or digits.
- Ensure error handling for invalid inputs, such as non-integer grades or   incorrect input format.

<font color = stream>Your solution should consist of a single Python file containing the code for this program. Remember to include comments to explain your code logic.</font>

```python
def students_details():
    name = input("Enter Student Name: ")
    maths = int(input('Enter Maths Marks: '))
    science = int(input('Enter Science Marks: '))
    english = int(input('Enter English Marks: '))
    return name, maths, science, english
    
def calc_average(maths, science, english):
    average = (maths + science + english) / 3
    return average

def performances(average_grade):
    if average_grade >= 90:
        return "Excellent"
    elif average_grade >= 80:
        return 'Good'
    elif average_grade >= 70:
        return "Average"
    else:
        return "Below Average"
        
def main():
    print('------------------WELCOME--------------------')
    print("========Grading System of ABC School=========")
    print('---------------------------------------------')
    print('---------------------------------------------')
    num_studs = int(input("Enter the number of students: "))
    student_result = []
    
    for i in range(num_studs):
        print('\nStudent', i+1, ":")
        name, maths, science , english = students_details()
        average_grade = calc_average(maths, science, english)
        performance = performances(average_grade)
        student_result.append((name, average_grade, performance))
        
    print("\nOutput:")
    print("Name      | Average Grade | Performance")
    print("----------------------------------------")
    for name, avg_grade, performance in student_result:
        print(f"{name:<10}| {avg_grade:.2f}{'' * (13 - len(str(avg_grade)))}         | {performance}")
              
if __name__ == '__main__':
              main()
```
